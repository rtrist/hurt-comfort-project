```python
import csv

from openai import OpenAI

client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

```

```python
system_setting = "You are a data analyst tasked with sorting tags from a csv file into these categories: Family/Friendship, Mental Health Issues, Romance, Sex, Violence, Other. You will print the percentage of each category in the tags you are given."
```

```python
import time

with open('tags_2007_2013.csv', 'r') as file:
    reader = csv.reader(file)
    all_tags = [row[0] for row in reader]

categories = ["Family/Friendship", "Mental Health Issues", "Romance", "Sex", "Violence", "Other"]

batch_size = 100  # Adjust as needed
delay_between_batches = 5  # Adjust as needed

# Loop through the tags in batches
for i in range(0, len(all_tags), batch_size):
    batch_tags = all_tags[i:i + batch_size]

    prompt = f"Given the tags: {', '.join(tags)}, categorize each tag into one of the following categories: {', '.join(categories)}. Calculate the percentage of each category present in the tags. Exclude tags that don't fall into any category."


    res = client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=[
                  {"role":"system", "content": system_setting},
                  {"role":"user", "content": prompt}
                ]
            )

    print(res.choices[0].message.content)

    time.sleep(delay_between_batches)
```
